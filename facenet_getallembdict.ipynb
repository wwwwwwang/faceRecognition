```python
import numpy as np
import tensorflow as tf

import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import sys
sys.path.append(r'D:\Users\Madhouse\tensorflow\yobiface\src')
import utils

import os
from os.path import join as pjoin
import sys
import copy
import shutil
import pickle

print(tf.__version__)
```

```python
tf.reset_default_graph()
sess = tf.Session()
saver = tf.train.import_meta_graph(r'D:\Users\Madhouse\tensorflow\download\20180402-114759\model-20180402-114759.meta')
saver.restore(sess, r'D:\Users\Madhouse\tensorflow\download\20180402-114759\model-20180402-114759.ckpt-275')
images_placeholder = tf.get_default_graph().get_tensor_by_name("input:0")
embeddings = tf.get_default_graph().get_tensor_by_name("embeddings:0")
phase_train_placeholder = tf.get_default_graph().get_tensor_by_name("phase_train:0")
```

```python
def load_data(data_dir, cnt=50, delete=False):
    data = {}
    pics_ctr = 0
    print(delete)
    for guy in os.listdir(data_dir):
        person_dir = pjoin(data_dir, guy)
        #print(person_dir)
        curr_pics = [utils.preproc(cv2.imread(pjoin(person_dir, f))) for f in os.listdir(person_dir)]
        #if len(curr_pics) < 2:
        #    continue
        data[guy] = curr_pics
        pics_ctr += len(curr_pics)
        if (delete):
            shutil.rmtree(person_dir)
        if len(data) == cnt:
            break
    return data, pics_ctr
```

```python
def rename_all_pics(data_dir):
    for guy in os.listdir(data_dir):
        person_dir = pjoin(data_dir, guy)
        print(person_dir)
        id=0
        #curr_pics = [utils.preproc(cv2.imread(pjoin(person_dir, f))) for f in os.listdir(person_dir)]
        #if len(curr_pics) < 2:
        #    continue
        for f in os.listdir(person_dir):
            os.rename(os.path.join(person_dir, f), os.path.join(person_dir, guy+"_"+str(id)) + ".jpg")
            id+=1
    print("finished...")

path = r'D:\software\myphotobase'
rename_all_pics(path)
```

```python
DATA_DIR = r'D:\software\lfw\lfw'

final_dict={}
for i in range((4069+100)//100):
    print("###i=",i)
    data, n_pics = load_data(DATA_DIR,100,True)
    print('Loaded %d pics.' % n_pics)
    emb_res = {}
    emb_dict = {}
    for guy in data:
        emb_dict[guy] = list(sess.run([embeddings], feed_dict={images_placeholder: np.array(data[guy]), phase_train_placeholder: False })[0])
        #print(emb_dict[guy])
        mean_embedding = np.mean(emb_dict[guy], axis=0)
        emb_res[guy]=mean_embedding
    final_dict.update(emb_res)
    print(len(final_dict))
```
